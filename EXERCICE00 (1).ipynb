```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as pyo
df=pd.read_csv('tips.csv')
df.head()
```

```python
df.describe()
```

#Figure 1

```python
df1=df[['total_bill','time','sex']]
df1.head()
```

```python
df1['c1']= '('+df1['time']+','+ df1['sex']+')'
df1=df1.drop(['time','sex'],axis=1)
df1.groupby('c1').sum()
df1.reset_index()
```

```python
fig = px.bar(df1, x="c1", y="total_bill")
fig.show()
```

#Figuire 2

```python
df2=pd.DataFrame({'total_bill':[df['total_bill'].mean()],'tip':[df['tip'].mean()],"size":[df['size'].mean()]})
```

```python
df2=df2.stack().reset_index().rename(columns={"level_1":"type", 0:"Value"})
```

```python
fig = px.bar(df2, x="type", y="Value")
fig.show()
```

#Figure 3

```python
df3=df[['total_bill','tip']]
df3.head()
```

```python
trace = go.Scatter(x=df['total_bill'], y=df['tip'], mode='markers')
data=[trace]
layout = go.Layout(title='correlation between total_bill and tips ',
                   xaxis=dict(title='total_bill'),
                   yaxis=dict(title='tip'))
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

#Figure 7

```python
df7=pd.read_csv('tips.csv',parse_dates=True)
df7=df7[['total_bill','day','time']]
df7=df7.groupby(['day','time']).sum().reset_index()
df7.head()
```
