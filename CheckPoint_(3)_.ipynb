```python
import pandas as pd
df=pd.read_csv('dataset.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Year'].fillna(df['Year'].median(),inplace=True)
```

```python
df['Publisher'].value_counts()
```

```python
df['Publisher'].fillna('Electronic Arts',inplace= True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
import plotly.express as px
fig = px.area(df, x='Year',y='Global_Sales',color='Publisher')
fig.show()
```

```python
fig = px.histogram(df, x='Global_Sales', nbins=20)
```

```python
fig.show()
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
trace = go.Scatter(x=df['EU_Sales'], y=df['NA_Sales'],mode='markers')
data=[trace]
layout = go.Layout(title='the relationship between North American sales and European sales',
                   xaxis=dict(title='EU_Sales'),
                   yaxis=dict(title='NA_Sales'))
fig2 = go.Figure(data=data, layout=layout)
pyo.plot(fig2)
```

```python
trace3 = go.Scatter(x=df['JP_Sales'], y=df['Other_Sales'],mode='markers')
data3=[trace3]
layout3 = go.Layout(title='the relationship between Japanese sales and Other sales',
                   xaxis=dict(title='JP_Sales'),
                   yaxis=dict(title='Other_Sales'))
fig3 = go.Figure(data=data3, layout=layout3)
pyo.plot(fig3)
```

```python
fig = px.area(df, x='Year',y='Global_Sales',color='Platform')
fig.show()
```

```python
df8=df[["Publisher","Global_Sales"]].groupby("Publisher").sum()
df8.head()
```

```python
df8=df8.sort_values('Global_Sales',ascending=False).iloc[0:10,:]
df8=df8.reset_index()
```

```python
fig = px.bar(df8, x="Publisher", y="Global_Sales", color="Publisher", height=400)
fig.show()
```
